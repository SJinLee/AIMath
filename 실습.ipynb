# 4. 공학적 도구로 추세선의 기울기 구하기(p120)
(https://www.algeomath.kr/)

1. 알지오매스 프로그램 실행하기
    * 손실함수 $L(a)= \frac{498a^2-996a+541}{10}$이 최솟값을 갖는 $a$의 값, 즉 최적 추세선의 기울기를 구하는 알지오매스 코드를 아래 주소에서 확인한다.
[http://me2.do/G2UDfgsP](http://me2.do/G2UDfgsP)

2. 자료 입력하기
    * 창의 2번째 블록에서 ‘연산 블록’을 클릭하고, 입력값과 측정값, 즉 [[3,4], [6,6.5], [9,10.5], [3,3], [6,3.5], [9,4.5], [5,4.5], [6,8.5], [8,7.5], [11,13.5]]를 입력한다.


3. 손실함수의 식 입력하기

    * 창 왼쪽의 $f(x)$ 탭을 클릭하고, 손실함수의 식 $L(a)= \frac{498a^2-996a+541}{10}$을 입력한다. 이때, 변수는 $a$ 대신 프로그램에 기본으로 설정되어 있는 변수인 $x$로 입력한다.


4. 학습률, 초깃값, 반복 횟수 입력하기
    * 창의 5, 6, 11번째 블록에서 '연산 블록'을 클릭하고, 학습률 r=0.001, 초깃값 a=3, 반복 횟수 100을 각각 입력한다. 이때 학습률은 'r' 대신 'alpha'로 설정되어 있음을 확인한다.
5. 새로운 초깃값을 정하는 식 입력하기
    * 창의 13번째 블록에서 ‘연산 블록’을 클릭하고, 새로운 초깃값을 정하는 식을 입력한다. 이때 $L'(a)= \frac{498}5 (a-1)$이므로 'a-alpha*(498/5)*(a-1)'을 입력한다.
6. 결과 확인하기
    * 창 하단의 버튼을 클릭하면 기하창에 결과가 나타난다. 이로부터 경사하강법을 적용하여 구한 추세선의 기울기는 1에 가까워짐을 알 수 있다. 따라서 최적 추세선은 $f(x)=x$이다

* Fully connected layer

image.png

```python
import tensorflow as tf
(x_train,y_train), (x_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
class Model(tf.keras.Model):
    def __init__(self):
        super(Model,self).__init__()
        initializer = tf.keras.initializers.glorot_uniform()
        self.w1 = tf.Variable(initializer([784,100]))
        self.b1 = tf.Variable(initializer([100]))
        self.w2 = tf.Variable(initializer([100,10]))
        self.b2 = tf.Variable(initializer([10]))

    def call(self,inputs): # input=(?x784)
        y = self.b1 + inputs @ self.w1
        y = tf.keras.activations.relu(y)
        y = self.b3 + y @ self.w3
        y = tf.keras.activations.sigmoid(y)
        return tf.squeeze(y)
```

```python
model = Model()
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.005),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(x_train.astype(np.float32).reshape([-1,784]),
          y_train.astype(np.float32),epochs=10)
```

```python
import pandas as pd
pd.DataFrame(history.history).plot()
```

```python
pred = model(x_test.astype(np.float32).reshape([-1,784]))
```

* Convolutional layer

```python
class Model(tf.keras.Model):
    def __init__(self):
        super(Model,self).__init__()
        self.conv1 = tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation="relu")
        self.pool1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))
        self.conv2 = tf.keras.layers.Conv2D(64, kernel_size=(3, 3), activation="relu")
        self.pool2 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))
        self.flatten = tf.keras.layers.Flatten()
        # self.dropout = tf.keras.layers.Dropout(0.5)
        self.dense3 = tf.keras.layers.Dense(10)
    def call(self,inputs):
        #y = inputs @ self.dense1.weights + self.dense1.bias
        y = self.conv1(inputs)
        y = self.pool1(y)
        y = self.conv2(y)
        y = self.pool2(y)
        y = self.flatten(y)
        y = self.dense3(y)
        y = tf.keras.activations.softmax(y)
        return y
```

```python
model = Model()
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.005),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(x_train.astype(np.float32).reshape([-1,28,28,1]),
          y_train.astype(np.float32),epochs=10)
```

```python
import pandas as pd
pd.DataFrame(history.history).plot()
```

```python
pred = model(x_test.astype(np.float32).reshape([-1,28,28,1]))
```

```python
# 숫자 그림을 위한 추가 함수
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = 'binary', **options)
    plt.axis("off")
```

```python
plt.figure(figsize=(9,9))
example_images = x_test[:100]
plot_digits(example_images, images_per_row=10)
plt.show()
```
